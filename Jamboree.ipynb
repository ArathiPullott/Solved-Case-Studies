## Multicollinearity check by VIF score(variables are dropped one-by-one till none has VIF>5)

Consider 'GRE Score' as dependent column

```python
def multi_colinear():
    vif_1st=[]
    for i in w_train.columns:
         x1=w_train[i].values
         x2=w_train.drop(i,axis=1)
         x2=x2.values
         model.fit(x2,x1)
         x1_hat=model.predict(x2)
         den=sum(((x1-x1.mean())**2))
         num=sum((x1-x1_hat)**2)
         score=1-(num/den)
         vif=1/(1-(score)**2)
         vif_1st.append(vif)
    return vif_1st
```

```python
multi_colinear()
```

VIF values of all columns has less then 5 it is not multi colinear
